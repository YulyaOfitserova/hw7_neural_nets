### Обзор Mask R-CNN

Архитектура напоминает Faster R-CNN с добавлением еще одной ветки, которая предсказывает положение маски, покрывающей найденный объект (instance segmentation).


n3cstpty6ktfwhw6vklswah1rxk.png

Предсказываются маски класса, не зная что изображено. А затем выбирается маска класса, победившего в независимом классификаторе. 
Одна из основных модификаций, возникших из-за необходимости предсказывать маску — изменение процедуры RoIPool(вычисляющей матрицу признаков для региона-кандидата) на так называемую RoIAlign.

Для вычисления точных значений признаков используется билинейная интерполяция по 4 ближайшим целочисленным точкам (используя максимальное или среднее значение) (см рис ниже). В отличие от RoIPool, где использовалось округление. 

ml27mqsszmzpo4-j0iv-yobw1os.png

Не важно расположение точки, главное в какой квадрат признаков она попала (синяя штриховка)

Плюсы, наверное, это скорость. Минусы - похожие по фооме обьекты, соответсвенно с близкими признаками, может выдавать за один класс.
